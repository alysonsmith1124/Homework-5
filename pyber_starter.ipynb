```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combo_df = pd.merge(ride_data, city_data, on = "city", how = "left")
# Display the data table for preview
combo_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Separate by each city type for color coordination in the graph
#Then group each type by the city names to find coordinate values
urban = combo_df[combo_df['type'] == 'Urban']
urb_group = urban.groupby(['city'])
u_xcoords = urb_group['ride_id'].count()
u_ycoords = urb_group['fare'].mean()
u_size = urb_group['driver_count'].mean()

suburban = combo_df[combo_df['type'] == 'Suburban']
sub_group = suburban.groupby(['city'])
s_xcoords = sub_group['ride_id'].count()
s_ycoords = sub_group['fare'].mean()
s_size = sub_group['driver_count'].mean()

rural = combo_df[combo_df['type'] == 'Rural']
rur_group = rural.groupby(['city'])
r_xcoords = rur_group['ride_id'].count()
r_ycoords = rur_group['fare'].mean()
r_size = rur_group['driver_count'].mean()

# Build the scatter plot for each city type to delineate color difference
urb_graph = plt.scatter(u_xcoords, u_ycoords, s = 10 * u_size, alpha = .70, edgecolors = "black", c = "Coral", linewidths = 1.2)

sub_graph = plt.scatter(s_xcoords, s_ycoords, s = 10 * s_size, alpha = .70, edgecolors = "black", c = "SkyBlue", linewidths = 1.2)

rur_graph = plt.scatter(r_xcoords, r_ycoords, s = 10 * r_size, alpha = .70, edgecolors = "black", c = "Gold", linewidths = 1.2)

# Incorporate the other graph properties
plt.grid()
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides Per City')
plt.ylabel('Average Fare Per City')

# Create a legend
lgnd = plt.legend((urb_graph, sub_graph, rur_graph), ('Urban','Suburban','Rural'), title = "City Types", scatterpoints = 1)
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(42, 33, 'Note:\nCircle size correlates with driver count per city.')
# Save Figure
plt.savefig("Bubbleplot.png")

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
grouped_fare_sum = combo_df.groupby(['type']).sum()['fare']
fare_sum = combo_df['fare'].sum()

fare_percent = 100 * grouped_fare_sum / fare_sum

# Build Pie Chart
plt.pie(fare_percent, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["Gold", "lightSkyBlue", "lightCoral"], 
        startangle=145, 
        autopct='%1.1f%%',
        explode=[0, 0, 0.1], 
        shadow=True)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("FarePieChart.png")

plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
grouped_ride_sum = combo_df.groupby(['type']).count()['ride_id']
ride_sum = combo_df['ride_id'].count()

ride_percent = 100 * grouped_ride_sum / ride_sum

# Build Pie Chart
plt.pie(ride_percent, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["Gold", "lightSkyBlue", "lightCoral"], 
        startangle=145, 
        autopct='%1.1f%%',
        explode=[0, 0, 0.1], 
        shadow=True)
plt.title("% of Total Ride by City Type")
# Save Figure
plt.savefig("RidePieChart.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
grouped_driver_sum = city_data.groupby(['type']).sum()['driver_count']
driver_sum = city_data["driver_count"].sum()

driver_percent = 100 * grouped_driver_sum / driver_sum

# Build Pie Chart
plt.pie(driver_percent, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["Gold", "lightSkyBlue", "lightCoral"], 
        startangle=145, 
        autopct='%1.1f%%',
        explode=[0, 0, 0.1], 
        shadow=True)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("DriversPieChart.png")
plt.show()

```

Analysis:
1) The larger the total number of rides per city, the cheaper the average fare and larger the driver count. With Urban cities leading the number of rides, with Suburban and Rural trailing respectively, the average fare and driver count follows this trend.
2) While the fares and number of rides for Urban cities are between 60-70%, the number of drivers is 81% of the total market. There may be an opportunity for Urban cities to increase the average fare per ride by capping the number of drivers in their cities. With a smaller supply, the demand will rise and prices will follow.
3) On the opposite side of the spectrum, Suburban cities have fares and number of rides between 25-35% of the total market with only 17% of total drivers. By increasing the number of drivers, the cost per ride should drop with the competition and the number of rides should increase. However, the practicality of this could be taken into further consideration due to the difference in distance between locations in Suburban vs. Urban cities.
